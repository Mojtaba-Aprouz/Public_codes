```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

# Fast Fourier transform and phase spectrum




```python
img_1=cv2.imread('/content/drive/MyDrive/computer_V/EX_1/bridge.tif')
img_1 = cv2.cvtColor(img_1,cv2.COLOR_BGR2GRAY)
dft_1 = np.fft.fft2(img_1)
dft_shift_1 = np.fft.fftshift(dft_1)
phase_1 = np.angle(dft_shift_1)

img_2=cv2.imread('/content/drive/MyDrive/computer_V/EX_1/tiger.tif')
img_2 = cv2.cvtColor(img_2,cv2.COLOR_BGR2GRAY)
dft_2 = np.fft.fft2(img_2)
dft_shift_2 = np.fft.fftshift(dft_2)
phase_2 = np.angle(dft_shift_2)


ax1 = plt.subplot(2,2,1)
ax1.imshow(img_1, cmap='gray')

ax2 = plt.subplot(2,2,2)
ax2.imshow(phase_1, cmap='gray')

ax3 = plt.subplot(2,2,3)
ax3.imshow(img_2, cmap='gray')

ax4 = plt.subplot(2,2,4)
ax4.imshow(phase_2, cmap='gray')


plt.show()
```
